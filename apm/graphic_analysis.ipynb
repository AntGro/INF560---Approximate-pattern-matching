# Get graphic representation of runtime performance

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from os import listdir
from os.path import isfile, join

mydir = "testsResults"

# Relative paths of files in dir @mypath
paths_file = [join(mydir, f) for f in listdir(mydir) if isfile(join(mydir, f))]
```

```python
results = []

for i in range(len(paths_file)):
    params = list(map(int,str.split(paths_file[i][paths_file[i].rfind("\\") + 1:].rstrip(".txt"))))
    file_res = [line.rstrip('\n') for line in open(paths_file[i])]
    for res in file_res:
        aux = params.copy()
        meth, t = str.split(res)
        aux.append(meth)
        aux.append(float(t))
        results.append(aux)

results = pd.DataFrame(results, columns=["n", "N", "dna_size", "nb_patterns", "method", "execution_time"])
```
